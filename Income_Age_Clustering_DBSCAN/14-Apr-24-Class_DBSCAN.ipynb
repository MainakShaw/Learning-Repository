```python
import pandas as pd
```

```python
data = pd.read_csv("income_age_clustering_data.csv")
```

```python
data2 = data[['Age','Income']]
```

```python
data2
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data2)
```

```python
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
```

```python
cluster_DB = DBSCAN(eps = 0.3, min_samples=12) #r => eps, M => min_samples
cluster_DB.fit(scaled_data)
print(cluster_DB.labels_)
plt.scatter(scaled_data[:,0],scaled_data[:,1],c = cluster_DB.labels_)
```

```python
import numpy as np
colormap = np.array(["red","blue","yellow", "black"])
plt.scatter(scaled_data[:,0],scaled_data[:,1],c = colormap[cluster_DB.labels_])
```

```python
from sklearn.metrics import silhouette_score
silhouette_score(data[['Age', 'Income']],cluster_DB.labels_)
```
