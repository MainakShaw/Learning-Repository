```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("credit_customers (1).csv")
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.drop('checking_status',axis = 1 , inplace =True)
```

```python
data
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
```

```python
set(data['credit_history'])
```

```python
data[['all paid',
 'critical/other existing credit',
 'delayed previously',
 'existing paid',
 'no credits/all paid']] = ohe.fit_transform(data[['credit_history']]).toarray()
```

```python
data
```

```python
data.drop('credit_history', axis = 1, inplace = True)
```

```python
data
```

```python
set(data['purpose'])
```

```python
data[[
    'business',
 'domestic appliance',
 'education',
 'furniture/equipment',
 'new car',
 'other',
 'radio/tv',
 'repairs',
 'retraining',
 'used car'
]] = ohe.fit_transform(data[['purpose']]).toarray()
```

```python
data.drop('purpose', axis = 1, inplace = True)
```

```python
data
```

```python
data.dtypes
```

```python
for column, dtype in zip(data.dtypes.index, data.dtypes.values):
    if dtype == "object" and column != "class":
        unique_values = set(data[column])
        unique_values = list(unique_values)
        data[unique_values] = ohe.fit_transform(data[[column]]).toarray()
        data.drop(column, inplace = True, axis = 1)
```

```python
data.dtypes
```

```python
X = data.drop('class', axis = 1) # apart from class column all columns will be used as IV
y = data['class'] # only class will be used as DV
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest =  train_test_split(X, y , test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_r = LogisticRegression()
```

```python
log_r.fit(xtrain,ytrain)
```

```python
ypred = log_r.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(ypred,ytest)
```

```python
# help (LogisticRegression)
```

```python
log_r = LogisticRegression(solver="liblinear")
log_r.fit(xtrain,ytrain)
ypred = log_r.predict(xtest)
accuracy_score(ypred,ytest)
```
