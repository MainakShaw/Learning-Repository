```python
from transformers import pipeline
classifier = pipeline("zero-shot-classification",
                      model="facebook/bart-large-mnli")
```

```python
sequence_to_classify = "I like this product very much!"
candidate_labels = ['negative review', 'positive review', 'neutral review']
classifier(sequence_to_classify, candidate_labels)

```

```python
sequence_to_classify = "I hated this product very much!"
candidate_labels = ['negative review', 'positive review', 'neutral review']
classifier(sequence_to_classify, candidate_labels)

```

```python
sequence_to_classify = "I am currently using Macbook."
candidate_labels = ['HP Laptop', 'Dell Laptop', 'Macbook Laptop']
classifier(sequence_to_classify, candidate_labels)

```

```python
sequence_to_classify = "I am currently working on Indutrial revolution topic."
candidate_labels = ['World War 2', 'Factories', 'World War']
classifier(sequence_to_classify, candidate_labels)

```

```python
sequence_to_classify = "I was working with Hitler."
candidate_labels = ['World War 2', 'Factories', 'World War 1']
classifier(sequence_to_classify, candidate_labels)

```
