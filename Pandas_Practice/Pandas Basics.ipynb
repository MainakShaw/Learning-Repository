## Filtering Data

```python
df=pokemonlist_cropped
df
```

```python
df.loc[(df["Type 1"]=='Grass') & (df["Type 2"]=='Dark')]
```

```python
df.loc[(df["Type 1"]=='Grass') | (df["Type 2"]=='Dark')]
```

```python
new_df=df.loc[((df["Type 1"]=='Grass') | (df["Type 2"]=='Dark')) & (df["HP"] >=80)]
new_df
```

```python
#Resetting the indexes for new_df
new_df=new_df.reset_index()
new_df
```

```python
#Resetting the indexes for new_df without keeping the old index column
new_df=df.loc[((df["Type 1"]=='Grass') | (df["Type 2"]=='Dark')) & (df["HP"] >=80)]
new_df=new_df.reset_index(drop=True)
new_df
# #doing it inplace(Conserves some memory)
# new_df=new_df.reset_index(drop=True,inplace=True)
# print(new_df)
```

```python
#doing it inplace(Conserves some memory)
new_df=df.loc[((df["Type 1"]=='Grass') | (df["Type 2"]=='Dark')) & (df["HP"] >=80)]
new_df.reset_index(drop=True,inplace=True)
new_df
```

```python
#Pokemons in new_df with names containing 'Mega'
new_df.loc[new_df["Name"].str.contains("Mega")]
```

```python
#Pokemons in new_df with names containing 'Mega'
new_df=df.loc[((df["Type 1"]=='Grass') | (df["Type 2"]=='Dark')) & (df["HP"] >=80)]
new_df.reset_index(drop=True,inplace=True)
new_df.loc[~new_df["Name"].str.contains("Mega")] #Tilde for not
```

```python
#The .str.contains() can also do complicated filtering using regex functions:-
import re
df.loc[df["Type 1"].str.contains("Fire|Grass",regex=True)]
```

```python
df.loc[df["Type 1"].str.contains("fire|grass",regex=True)] #Case matters
```

```python
df.loc[df["Type 1"].str.contains("fire|grass",flags=re.I,regex=True)] #Regex flag ignorecase
```

```python
df.loc[df["Name"].str.contains("pi[a-z]*",flags=re.I,regex=True)]
```

```python
df.loc[df["Name"].str.contains("^pi[a-z]*",flags=re.I,regex=True)]
```

## Working with large amounts of Data

```python
new_df=pd.DataFrame(columns=df.columns)
for df in pd.read_csv("modified_pokemon_data.csv",chunksize=5):
    results=df.groupby(["Type 1"]).count()
    new_df=pd.concat([new_df,results])
```
