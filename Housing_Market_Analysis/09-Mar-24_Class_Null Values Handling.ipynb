# Treatment of null values

```python
data= pd.read_csv("Housing Data.csv")
```

```python
data
```

```python
data.columns
```

```python
data.Frontage.isnull().sum()
```

```python
Test_column = data[['Frontage']]
```

```python
Test_column
```

```python
Test_column['Frontage'].mean()
```

```python
Test_column['Frontage_mean'] = Test_column['Frontage'].fillna(Test_column['Frontage'].mean())
```

```python
Test_column
```

```python
Test_column.isnull().sum()
```

```python
Test_column['Frontage'].median()
```

```python
Test_column['Frontage_median'] = Test_column['Frontage'].fillna(Test_column['Frontage'].median())
```

```python
Test_column
```

```python
Test_column['Frontage'].mode()
```

```python
type(Test_column['Frontage'].mode())
```

```python
Test_column['Frontage'].mode()[0]
```

```python
Test_column['Frontage_mode'] = Test_column['Frontage'].fillna(Test_column['Frontage'].mode()[0])
```

```python
Test_column
```

```python
Test_column['Frontage_ffill'] = Test_column['Frontage'].fillna(method = "ffill")#Only use when the first value in the column is not null
```

```python
Test_column
```

```python
Test_column['Frontage_bfill'] = Test_column['Frontage'].fillna(method = "bfill")#Only use when the last value in the column is not null
```

```python
Test_column
```

```python
Test_column['Frontage_interpolate'] = Test_column['Frontage'].interpolate() #Use when both the first and last value are null for a column
#interpolate() replaaces the null by the average of the previous value and the vale afterwards
```

```python
Test_column
```

```python
Test_column.describe()#Chgeck for std vale closest to that of the original column, pick that method to replace the null values(most suitable)
```

```python
# intrerpolate method is the best method to handle null values for frontage column
```
