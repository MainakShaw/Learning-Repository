# Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
x=[i for i in range(101)]
y=[i**2 for i in range(101)]
plt.plot(x,y,label="x^2",color="red",linewidth=2, marker=".", markersize=1, markeredgecolor="black", linestyle="--")
plt.title("Our First Graph:-",fontdict={"family":"serif","fontsize":20})
plt.xlabel("X Axis:-",fontdict={"family":"serif","fontsize":20})
plt.ylabel("Y Axis:-",fontdict={"family":"serif","fontsize":20})
plt.xticks([i for i in range(10,101,10)])
plt.yticks([i**2 for i in range(10,101,10)])
plt.legend()
plt.show()
```

```python
#Shorthand notation for plotting pretty much the same thing
## Shorthand Notation:-
## fmt=[color][marker][line]
x=[i for i in range(101)]
y=[i**2 for i in range(101)]
plt.plot(x,y,"8-.c",label="x^2",linewidth=1,markersize=3)
plt.title("Our Second Graph:-",fontdict={"family":"serif","fontsize":20})
plt.xlabel("X Axis:-",fontdict={"family":"serif","fontsize":20})
plt.ylabel("Y Axis:-",fontdict={"family":"serif","fontsize":20})
plt.xticks([i for i in range(10,101,10)])
plt.yticks([i**2 for i in range(10,101,10)])
plt.legend()
plt.show()
```

```python
#Adding another function in the same graph:-
x1=np.arange(0,11,0.5)
y1=x1**2
y2=x1**3

plt.plot(x1,y1,"r-",label="x^2",linewidth=2, marker=".", markersize=1, markeredgecolor="black")
plt.plot(x1,y1,"r--",label="x^2",linewidth=2, marker=".", markersize=1, markeredgecolor="black")

plt.plot(x1,y2,"g-",label="x^3",linewidth=2, marker=".", markersize=1, markeredgecolor="blue")
plt.plot(x1,y2,"g--",label="x^3",linewidth=2, marker=".", markersize=1, markeredgecolor="blue")

plt.title("Our Third Graph:-",fontdict={"family":"serif","fontsize":20})
plt.xlabel("X Axis:-",fontdict={"family":"serif","fontsize":20})
plt.ylabel("Y Axis:-",fontdict={"family":"serif","fontsize":20})
plt.xticks([i for i in range(0,11,1)])
plt.yticks([i for i in range(0,1001,100)])
plt.legend()
plt.show()

```

```python
#How to use different styles for same graph, breaking it up into smaller sections
```

```python
x1=np.arange(0,11,0.5)

plt.plot(x1[:15],x1[:15]**2,"r-",label="x^2",linewidth=2, marker=".", markersize=1, markeredgecolor="black")
plt.plot(np.arange(7.0,8.1,0.1),np.arange(7.0,8.1,0.1)**2,"r--",linewidth=2)
plt.plot(x1[15:],x1[15:]**2,"r--",label="x^2",linewidth=2, marker=".", markersize=1, markeredgecolor="black")

plt.plot(x1[:15],x1[:15]**3,"g-",label="x^3",linewidth=2, marker=".", markersize=1, markeredgecolor="blue")
plt.plot(np.arange(7.0,8.1,0.1),np.arange(7.0,8.1,0.1)**3,"g--",linewidth=2)
plt.plot(x1[15:],x1[15:]**3,"g--",label="x^3",linewidth=2, marker=".", markersize=1, markeredgecolor="blue")

plt.title("Our 4th Graph:-",fontdict={"family":"serif","fontsize":20})
plt.xlabel("X Axis:-",fontdict={"family":"serif","fontsize":20})
plt.ylabel("Y Axis:-",fontdict={"family":"serif","fontsize":20})
plt.xticks([i for i in range(0,11,1)])
plt.yticks([i for i in range(0,1001,100)])
plt.legend()
plt.show()
```

```python
#Resizing your graph
plt.figure(figsize=(5,3),dpi=300)
#figsize=Ratio of x-axis to y-axis (tuple values are in inches)
#dpi=number of pixels per inch(use larger number for bigger image)
x1=np.arange(0,11,0.5)

plt.plot(x1[:15],x1[:15]**2,"r-",label="x^2",linewidth=2, marker=".", markersize=3, markeredgecolor="black")
plt.plot(np.arange(7.0,8.1,0.1),np.arange(7.0,8.1,0.1)**2,"r--",linewidth=2)
plt.plot(x1[15:],x1[15:]**2,"r--",label="x^2",linewidth=2, marker=".", markersize=3, markeredgecolor="black")

plt.plot(x1[:15],x1[:15]**3,"g-",label="x^3",linewidth=2, marker=".", markersize=3, markeredgecolor="blue")
plt.plot(np.arange(7.0,8.1,0.1),np.arange(7.0,8.1,0.1)**3,"g--",linewidth=2)
plt.plot(x1[15:],x1[15:]**3,"g--",label="x^3",linewidth=2, marker=".", markersize=3, markeredgecolor="blue")

plt.title("Our 5th Graph:-",fontdict={"family":"serif","fontsize":20})
plt.xlabel("X Axis:-",fontdict={"family":"serif","fontsize":20})
plt.ylabel("Y Axis:-",fontdict={"family":"serif","fontsize":20})
plt.xticks([i for i in range(0,11,1)])
plt.yticks([i for i in range(0,1001,100)])
plt.legend()
plt.show()
```

## Bar Charts:-

```python
labels=["A","B","C"]
values=[1,2,3]
plt.bar(labels,values)
plt.figure(figsize=(3,5)) #and we can pass parametrs like figsize here as well
plt.show()
```

```python
#Setting hatch for various bars
labels=["A","B","C"]
values=[1,2,3]
bars=plt.bar(labels,values)

bars[0].set_hatch("/")
bars[1].set_hatch("O")
bars[2].set_hatch("+*")

plt.figure(figsize=(3,5))
plt.show()
```

```python
#Setting hatch for various bars(in a loop)
labels=["A","B","C"]
values=[1,2,3]
bars=plt.bar(labels,values)

list1=["//","O","+*"]
for bar in bars:
    bar.set_hatch(list1.pop(0))

plt.figure(figsize=(3,5))
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.title("Gas Prices over time in USD",fontdict={"fontweight":"bold","fontsize":22})
for country in gas.columns:
    if country != "Year":
        plt.plot(gas.Year,gas[country], marker=".")
plt.xticks(gas["Year"][::3].tolist()+[2011])
plt.xlabel("Year")
plt.ylabel("US Dollars")
plt.legend(gas.columns[1:],bbox_to_anchor=(1.0,1.0))
plt.savefig("Gas Prices Figure.png",dpi=300,bbox_inches="tight")
plt.show()
```

```python
#another way of doing it
weights_arr=np.array([float(x.strip("lbs")) if type(x)==str else float(x) for x in fifa["Weight"]])
weights_arr_in_kg=weights_arr*(0.45359237)
```

```python
weights_arr_in_kg
```

```python
fifa["Weight in kg"]=weights_arr_in_kg
fifa.head(5)
```

```python
for i,x in enumerate(fifa.columns):
    if x == "Weight":
        print(i)
```

```python
cols=list(fifa.columns.values)
len(cols)
```

```python
fifa=fifa[cols[0:28]+[cols[-1]]+cols[28:-1]]
fifa.columns.values[25:31]
```

```python
#Setting up weight categories
light=fifa.loc[fifa["Weight in kg"]<(125*0.45359237)].count()[0]
light_medium=fifa.loc[(fifa["Weight in kg"]>=(125*0.45359237)) & (fifa["Weight in kg"]<(150*0.45359237))].count()[0]
medium=fifa.loc[(fifa["Weight in kg"]>=(150*0.45359237)) & (fifa["Weight in kg"]<(175*0.45359237))].count()[0]
medium_heavy=fifa.loc[(fifa["Weight in kg"]>=(175*0.45359237)) & (fifa["Weight in kg"]<(200*0.45359237))].count()[0]
heavy=fifa.loc[fifa["Weight in kg"]>=(200*0.45359237)].count()[0]
print(f"Light = {light}")
print(f"Medium Light = {light_medium}")
print(f"Medium = {medium}")
print(f"Medium Heavy = {medium_heavy}")
print(f"Heavy = {heavy}")
```

```python
weight_categories=[light,light_medium,medium,medium_heavy,heavy]
weight_category_names=["Light(<56.699 kgs)","Medium Light(56.699 kgs-68.0389 kgs)","Medium(68.0389 kgs-79.3787 kgs)","Medium Heavy(79.3787 kgs-90.7185 kgs)","Heavy(>=90.7185 kgs)"]
plt.figure(dpi=200)
plt.style.use('bmh')
plt.pie(weight_categories,autopct="%.2f %%",labels=weight_category_names, radius=1.5)
plt.show()
```

```python
weight_categories=[light,light_medium,medium,medium_heavy,heavy]
weight_category_names=["Light(<56.699 kgs)","Medium Light(56.699 kgs-68.0389 kgs)","Medium(68.0389 kgs-79.3787 kgs)","Medium Heavy(79.3787 kgs-90.7185 kgs)","Heavy(>=90.7185 kgs)"]
explode_values=[0.8,0.02,0.02,0.02,0.8]
plt.figure(dpi=200)
plt.style.use('bmh')
plt.title("Weight Distribution of Fifa Players(in Kgs):-",y=1.15)
plt.pie(weight_categories,autopct="%.2f %%",labels=weight_category_names, radius=1.5, pctdistance=0.5, explode=explode_values)
plt.show()
```

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

weather_data=pd.read_csv("Weather_Data.csv")
corr_matrix = weather_data.corr()


plt.figure(figsize=(15,13),dpi=300)

# Calculate correlation between columns
corr_matrix = weather_data.corr()
values_matrix=np.array(corr_matrix.iloc[:,:])

# Plot the heatmap
plt.imshow(corr_matrix, cmap="magma")
for i in range(len(values_matrix)):
    for j in range(len(values_matrix)):
        if values_matrix[i, j] > 0.3:
            text = plt.text(j, i, '%.2f' % values_matrix[i, j],
                           ha="center", va="center", color="b")
        else:
            text = plt.text(j, i, '%.2f' % values_matrix[i, j],
                           ha="center", va="center", color="w")

plt.colorbar()
 
# Set plot title and axis labels
plt.title("Correlation Matrix Of The Dataset")
plt.xlabel("Features")
plt.ylabel("Features")
 
# Set tick labels
plt.xticks(range(len(corr_matrix.columns)),corr_matrix.columns, rotation=90)
plt.yticks(range(len(corr_matrix.columns)),corr_matrix.columns)
 
# Display the plot
plt.show()
```
