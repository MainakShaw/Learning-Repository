```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("FuelConsumption.csv")
```

```python
data.head()
```

```python
data.columns
```

```python
X = data[['ENGINESIZE']] #IV Should be always 2D
```

```python
y = data['CO2EMISSIONS'] # dv
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest= train_test_split(X, y, test_size=0.2) 
```

```python
xtrain.shape, ytrain.shape
```

```python
xtest.shape, ytest.shape
```

# Training of Machine learning model

```python
from sklearn.linear_model import LinearRegression
```

```python
linReg = LinearRegression()
```

```python
# for training model we use .fit method 
linReg.fit(xtrain, ytrain)
```

```python
# for testing we use .predict method
ypred = linReg.predict(xtest)
```

```python
# to check accuracy compare (ypred, ytest)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(ypred,ytest)
```

```python
new_truck = [[3.7]]
```

```python
linReg.predict(new_truck)
```

```python
plt.scatter(xtest,ytest, c = 'b', label = "Data points")
plt.plot(xtest, linReg.coef_[0] * xtest + linReg.intercept_, c = 'r', label = "Regression Line")
plt.xlabel("Engine Size")
plt.ylabel("Co2Emission")
plt.legend()
```

```python
m = linReg.coef_[0]
c = linReg.intercept_
```

```python
plt.scatter(xtest,ytest, c = 'b', label = "Data points")
plt.plot(xtest, m * xtest + c, c = 'r', label = "Regression Line")
plt.xlabel("Engine Size")
plt.ylabel("Co2Emission")
plt.legend()
```
