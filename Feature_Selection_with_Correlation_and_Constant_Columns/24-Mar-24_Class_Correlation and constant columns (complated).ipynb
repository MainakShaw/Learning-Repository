```python
from sklearn.feature_selection import VarianceThreshold
```

```python
var_T = VarianceThreshold()
```

```python
var_T.fit(data)
```

```python
len(var_T.get_support())
```

```python
var_T.get_support()
```

```python
data.columns[var_T.get_support() == False]
```

```python
data['AMT_CREDIT'].hist()
```

```python
set(data['FLAG_DOCUMENT_7'])
```

```python
drop_columns = data.columns[var_T.get_support() == False]
```

```python
data.drop(drop_columns, axis = 1, inplace=True)
```
