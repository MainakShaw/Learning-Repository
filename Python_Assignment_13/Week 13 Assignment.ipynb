```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import accuracy_score, silhouette_score
```

```python
X,y = make_blobs(n_samples=800,n_features=2,centers=4,cluster_std=[1.5,1,0.9,3],shuffle=True)
```

```python
X.shape
```

```python
set(y) #Unique values in y
```

```python
y.shape
```

### Checking for KMeans with n_clusters=2

```python
Kmeans = KMeans(n_clusters=2,n_init="auto")
Kmeans.fit(X)
```

```python
Kmeans.labels_
```

```python
silhouette_score(X,Kmeans.labels_)
```

```python
plt.scatter(X[:,0], X[:,1], c = Kmeans.labels_)
plt.plot()
```

### Checking for KMeans with n_clusters=3

```python
Kmeans = KMeans(n_clusters=3,n_init="auto")
Kmeans.fit(X)
```

```python
Kmeans.labels_
```

```python
silhouette_score(X,Kmeans.labels_)
```

```python
plt.scatter(X[:,0], X[:,1], c = Kmeans.labels_)
plt.plot()
```

### Checking for KMeans with n_clusters=4

```python
Kmeans = KMeans(n_clusters=4,n_init="auto")
Kmeans.fit(X)
```

```python
Kmeans.labels_
```

```python
silhouette_score(X,Kmeans.labels_)
```

```python
plt.scatter(X[:,0], X[:,1], c = Kmeans.labels_)
plt.plot()
```

```python
#Scaling the Features
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

### Checking for DBSCAN with eps=0.6 and min_samples=20

```python
cluster_DB = DBSCAN(eps = 0.6, min_samples=20)
cluster_DB.fit(X_scaled)
print(cluster_DB.labels_)
```

```python
plt.scatter(X_scaled[:,0],X_scaled[:,1],c = cluster_DB.labels_)
```

```python
silhouette_score(X_scaled,cluster_DB.labels_)
```

### Checking for DBSCAN with eps=0.5 and min_samples=30

```python
cluster_DB = DBSCAN(eps = 0.5, min_samples=30)
cluster_DB.fit(X_scaled)
print(cluster_DB.labels_)
```

```python
plt.scatter(X_scaled[:,0],X_scaled[:,1],c = cluster_DB.labels_)
```

```python
silhouette_score(X_scaled,cluster_DB.labels_)
```

### Checking for DBSCAN with eps=0.4 and min_samples=50

```python
cluster_DB = DBSCAN(eps = 0.4, min_samples=50)
cluster_DB.fit(X_scaled)
print(cluster_DB.labels_)
```

```python
plt.scatter(X_scaled[:,0],X_scaled[:,1],c = cluster_DB.labels_)
```

```python
silhouette_score(X_scaled,cluster_DB.labels_)
```

### Checking for DBSCAN with eps=0.4 and min_samples=60

```python
cluster_DB = DBSCAN(eps = 0.4, min_samples=60)
cluster_DB.fit(X_scaled)
print(cluster_DB.labels_)
```

```python
plt.scatter(X_scaled[:,0],X_scaled[:,1],c = cluster_DB.labels_)
```

```python
silhouette_score(X_scaled,cluster_DB.labels_)
```

1. Kmeans as a clustering algorithm can only create circular clusters. And it can't detect outliers. It has issues with data points overlapping between different clusters.
2. DBSCAN on the other hand seperates datapoints between core points,boundary points and outliers, based on the radius epsilon and the minimum no. of samples within that radius for each data point. Lastly the closest core points and their corresponding boundary points are grouped into clusters. This fixes the issues which we have with KMeans.
